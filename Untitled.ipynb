```python
preference_functions = {
    "Prix": "min",
    "Vitesse_Max": "max",
    "Conso_moy": "min",
    "Dis_freinage": "min",
    "Confort": "max",
    "Vol_Coffre": "max",
    "Acceleration": "max"
}
```

```python
normalized_df = (data_voiture - data_voiture.min()) / (data_voiture.max() - data_voiture.min())
```

```python
normalized_df
```

```python
# Calculer les indices PROMETHEE (P et D)
P_indices = np.zeros((len(data_voiture), len(data_voiture)))
D_indices = np.zeros((len(data_voiture), len(data_voiture)))
```

```python
D_indices
```

```python
for i in range(len(data_voiture)):
    for j in range(len(data_voiture)):
        if i != j:
            for critere, preference in preference_functions.items():
                if preference == "min":
                    P_indices[i, j] += np.maximum(0, normalized_df.iloc[i][critere] - normalized_df.iloc[j][critere])
                elif preference == "max":
                    P_indices[i, j] += np.maximum(0, normalized_df.iloc[j][critere] - normalized_df.iloc[i][critere])
                D_indices[i, j] += np.abs(normalized_df.iloc[i][critere] - normalized_df.iloc[j][critere])
```

```python
P_global = P_indices.sum(axis=1) - P_indices.sum(axis=0)
D_global = D_indices.sum(axis=1) - D_indices.sum(axis=0)
```

```python
# Classement final
scores = P_global - D_global
data_voiture["Score"] = scores
data_voiture = data_voiture.sort_values(by="Score", ascending=False)
```

```python
import numpy as np
import pandas as pd

def process_data(data, poids):
    nb_candidats = len(data)
    comparaison_matrice = np.zeros((nb_candidats, nb_candidats))

    # Compare les candidats deux à deux en conservant les poids pertinents
    for i in range(nb_candidats):
        for j in range(nb_candidats):
            if i != j:
                for k in range(len(poids)):
                    if data.iloc[i, k] >= data.iloc[j, k]:
                        comparaison_matrice[i, j] += poids[k]

    print("Matrice de comparaison des candidats :")
    print(comparaison_matrice)

    flux_positif = comparaison_matrice.sum(axis=0)
    flux_negatif = comparaison_matrice.sum(axis=1)

    # Récupérer les noms des candidats à partir des premières valeurs de chaque ligne
    candidat_names = [data.iloc[i, 0] for i in range(nb_candidats)]

    flux_dict_po = {candidat_names[i]: flux for i, flux in enumerate(flux_positif)}
    print("Dictionnaire des valeurs de flux associées aux candidats (Flux Positif):")
    print(flux_dict_po)

    flux_dict_neg = {candidat_names[i]: flux for i, flux in enumerate(flux_negatif)}
    print("Dictionnaire des valeurs de flux associées aux candidats (Flux Négatif):")
    print(flux_dict_neg)

    sorted_candidats_po = sorted(flux_dict_po, key=lambda candidat: flux_dict_po[candidat], reverse=True)
    print("Ordre des candidats en fonction des valeurs de flux positif (ordre décroissant):")
    for i, candidat in enumerate(sorted_candidats_po, start=1):
        print(f"Position {i}: {candidat}")

    sorted_candidats_neg = sorted(flux_dict_neg, key=lambda candidat: flux_dict_neg[candidat])
    print("Ordre des candidats en fonction des valeurs de flux négatif (ordre croissant):")
    for i, candidat in enumerate(sorted_candidats_neg, start=1):
        print(f"Position {i}: {candidat}")

# Exemple d'utilisation avec des données Pandas
data = pd.DataFrame({
    'Candidat': ['A1', 'A2', 'A3', 'A4', 'A5', 'A6'],
    'C1': [16, 10, 18, 18, 16, 6],
    'C2': [14, 18, 12, 4, 10, 14],
    'C3': [16, 12, 6, 20, 12, 18]
})

poids = [0.6, 0.1, 0.3]

process_data(data, poids)
```
